# Maximal Frequent Itemsets
This notebook identifies maximal frequent itemsets from Apriori results.

```python
import pandas as pd

# Load the frequent itemsets
frequent_itemsets = pd.read_csv('frequent_itemsets.csv', converters={'itemsets': eval})
frequent_itemsets['itemsets'] = frequent_itemsets['itemsets'].apply(frozenset)
```

```python
# Identify maximal itemsets
maximal_itemsets = []
for i, row_i in frequent_itemsets.iterrows():
    is_maximal = True
    for j, row_j in frequent_itemsets.iterrows():
        if row_i['itemsets'] < row_j['itemsets']:
            is_maximal = False
            break
    if is_maximal:
        maximal_itemsets.append(row_i)

maximal_df = pd.DataFrame(maximal_itemsets)
maximal_df.to_csv("maximal_frequent_itemsets.csv", index=False)
maximal_df.head()
```
