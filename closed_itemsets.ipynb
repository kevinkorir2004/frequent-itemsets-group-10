# Closed Frequent Itemsets
This notebook identifies closed frequent itemsets from Apriori results.

```python
import pandas as pd

# Load the frequent itemsets
frequent_itemsets = pd.read_csv('frequent_itemsets.csv', converters={'itemsets': eval})
frequent_itemsets['itemsets'] = frequent_itemsets['itemsets'].apply(frozenset)
```

```python
# Identify closed itemsets
closed_itemsets = []
for i, row_i in frequent_itemsets.iterrows():
    is_closed = True
    for j, row_j in frequent_itemsets.iterrows():
        if row_i['itemsets'] < row_j['itemsets'] and row_i['support'] == row_j['support']:
            is_closed = False
            break
    if is_closed:
        closed_itemsets.append(row_i)

closed_df = pd.DataFrame(closed_itemsets)
closed_df.to_csv("closed_frequent_itemsets.csv", index=False)
closed_df.head()
```
